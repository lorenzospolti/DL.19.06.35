```python
import os
import glob
import pickle
import pandas as pd

!gdown --id 1oi3mnz6PQVt-tEMR6IQnqC0ab9IZ1iXx

# Specify the path to the pickle file
pickle_file_path = '/content/extracted_information.pkl'

# Load the data from the pickle file into a list of dictionaries
with open(pickle_file_path, 'rb') as file:
    extracted_data = []
    while True:
        try:
            document = pickle.load(file)
            extracted_data.extend(document)
        except EOFError:
            break

# Convert the list of dictionaries to a Pandas DataFrame
df = pd.DataFrame(extracted_data)

# Now 'df' is a Pandas DataFrame that contains the extracted information from the XML files.
# You can use standard Pandas DataFrame operations to analyze and manipulate the data.
# For example, you can print the first few rows of the DataFrame:
display(df.head())
```

## Queries & Qrels

```python
import xml.etree.ElementTree as ET
# Queries
!gdown --id 1Nd_5jOEtazc8UH5Fp1oHb-coVLwas_cj
path_to_queries = 'topics2021.xml'

with open(path_to_queries, "r", encoding="utf8") as f:
    xml_text = f.read()

# Parse the XML data
root = ET.fromstring(xml_text)

# Extract data from XML and create a list of dictionaries
data = []
for topic in root.findall('.//topic'):
    topic_number = topic.get('number')
    text = topic.text.strip()
    data.append({'qid': topic_number, 'query': text})

# Create a DataFrame from the list of dictionaries
queries = pd.DataFrame(data)

# Display the DataFrame
display(queries)


#Qrels
!gdown --id 169ClkKps6wWud6YSOhNgKd4wi4gefyRN
path_to_qrels = 'clc_qrels2021.txt'
qrels = pd.read_csv(path_to_qrels, names=['qid','Q0','docno','label'],sep=" ",header=None)
qrels = qrels.drop(columns=['Q0'])
qrels["qid"] = qrels["qid"].astype(str)
qrels["docno"] = qrels["docno"].astype(str)
display(qrels.head(2))
```

```python
import re
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')

# We will keep the original query as full narrative
queries["query_full"] = queries["query"].astype(str)

# Patterns to detect sentences that talk about the current / recent situation
CURRENT_PATTERNS = [
    r'\bnow\b', r'\bcurrently\b', r'\btoday\b', r'\brecently\b',
    r'\bwas recently diagnosed\b', r'\bpresents with\b', r'\bis admitted\b',
    r'\bcomplains of\b', r'\bhas been\b', r'\bis being treated\b',
    r'\bon examination\b', r'\bongoing\b'
]
compiled_patterns = [re.compile(pat, re.IGNORECASE) for pat in CURRENT_PATTERNS]

def current_score(sentence: str) -> int:
    return sum(bool(p.search(sentence)) for p in compiled_patterns)

def extract_current_sentences(text: str, k: int = 3) -> str:
    sentences = nltk.sent_tokenize(text)
    if not sentences:
        return text
    scored = [(s, current_score(s)) for s in sentences]
    scored.sort(key=lambda x: x[1], reverse=True)
    # keep sentences with non-zero score, up to k
    filtered = [s for s, sc in scored if sc > 0][:k]
    if not filtered:
        # fallback: first k sentences
        filtered = sentences[:k]
    return " ".join(filtered)

# Build a "current situation" version of each query
queries["query_current"] = queries["query_full"].apply(extract_current_sentences)

# Build a combined query that gives more weight to current situation
def combine_queries(row):
    # repeat current situation to increase its impact in BM25
    return (row["query_current"] + " ") * 2 + row["query_full"]

queries["query_combined"] = queries.apply(combine_queries, axis=1)

display(queries.head())
```

```python
# --- Build a better index with field weighting ---

def iter_docs():
    for _, row in df.iterrows():
        title = str(row.get("official_title", ""))
        cond  = str(row.get("condition", ""))
        elig  = str(row.get("eligibility_criteria", ""))

        # Simple weighting trick: repeat important fields
        parts = [
            (title + " ") * 3,   # title x3
            (cond  + " ") * 2,   # condition x2
            elig                  # eligibility criteria x1
        ]
        # Add optional extra fields once
        for col in extra_fields:
            parts.append(str(row.get(col, "")))

        text = " ".join(parts).strip()

        yield {
            "docno": str(row["nct_id"]),  # or whichever ID column you have
            "text": text
        }

index_path = "./ct_index"
if os.path.exists(index_path):
    shutil.rmtree(index_path)
os.makedirs(index_path, exist_ok=True)

# Build the index using the IterDictIndexer signature
indexer = pt.IterDictIndexer(
    index_path,
    meta={"docno": 20},         # store docno as metadata (up to 20 chars)
    text_attrs=["text"],        # which field(s) contain the text
    meta_reverse=["docno"],     # enable reverse lookup on docno
    pretokenised=False,
    fields=False,
    threads=1,
)

index_ref = indexer.index(iter_docs())

# Open the index to ensure it is valid
index = pt.IndexFactory.of(index_ref)

# Print a simple summary
print("Index location:", index_path)
print("Indexed documents:", index.getCollectionStatistics().getNumberOfDocuments())

# Printing the files related to the index
print("Index files:")
!ls -lh ./ct_index
```

```python
# Retrieve collection statistics
stats = index.getCollectionStatistics()

print("Terrier Collection Statistics")
print("--------------------------------")
print(f"Indexed documents:        {stats.getNumberOfDocuments()}")
print(f"Unique terms (vocabulary): {stats.getNumberOfUniqueTerms()}")
print(f"Total tokens:             {stats.getNumberOfTokens()}")
print(f"Average document length:  {stats.getAverageDocumentLength():.2f}")
```

```python
from pyterrier.measures import *

# Choose which query representation to use in the experiment:
# Try "query_full", "query_current", and "query_combined" separately if you want.
queries_eval = queries[["qid", "query_combined"]].rename(columns={"query_combined": "query"})

retrievers = []
names = []

# Baseline TF-IDF
tfidf = pt.terrier.Retriever(index, wmodel="TF_IDF")
retrievers.append(tfidf)
names.append("TF_IDF")

# BM25 grid search over a few k1 and b values
for k1 in [0.8, 1.2, 1.6, 2.0]:
    for b in [0.3, 0.6, 0.9]:
        bm = pt.terrier.Retriever(
            index,
            wmodel="BM25",
            controls={"k_1": k1, "b": b}
        )
        retrievers.append(bm)
        names.append(f"BM25_k1={k1}_b={b}")

# Run the experiment – key metric: P(rel=2)@10
pt.Experiment(
    retrievers,
    queries_eval,
    qrels,
    eval_metrics=["map", "P_2@10"],
    names=names
)
```
